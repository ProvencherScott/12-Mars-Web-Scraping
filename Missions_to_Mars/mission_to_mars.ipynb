```python
### Mars Hemispheres - Scrape the images
# The URL provided:https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars
```

```python
# URL of page to be scraped
# hemisphere_url https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars
hemisphere_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
# access url
browser.visit(hemisphere_url)
```

```python
# Scrape page into soup
html = browser.html
hemi_soup = BeautifulSoup(html,'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(hemi_soup.prettify())
```

```python
# Finding links images for each hemisphere

# Create empty list to store image urls
links = []

# Locate image links
urls = hemi_soup.find_all("div", class_="item")

# Add base link to image links and append to the list
for url in urls:
    hemis_links = url.find('a')['href']
    path = 'https://astrogeology.usgs.gov' + hemis_links
    links.append(path)

links
```

```python
# Need to add url link to the image source
# Finding urls to full sized images for each hemisphere
image_url = []

# Click each image link 
for link in links:
    browser.visit(link)
    html = browser.html
    hemi_soup = BeautifulSoup(html, 'html.parser')
    #hemisphere_soup = BeautifulSoup(response.content,'html.parser')
    
    
# Find full jpgs
    url = hemi_soup.find_all('img', class_='thumb')
    each_url = url[0]['src']

# Add base link to image links and append to the list
    final_path = 'https://astrogeology.usgs.gov' + each_url
    image_url.append(final_path)

image_url
```

```python
# Find the names of the hemispheres
hemisphere_list = hemi_soup.find_all('div', class_='description')
hemisphere_list
```

```python
# Finding the hemisphere names

# Create an empty list to store hemisphere names
hemisphere_names = []

# Click each image link 
for link in links:
    browser.visit(link)
    html = browser.html
    hemi_soup = BeautifulSoup(html, 'html.parser')
    
    #hemisphere_soup = BeautifulSoup(response.content,'html.parser')
# Search for image titles
hemispheres = hemi_soup.find_all('div', class_='container')
names = hemispheres[0].find_all("h3")

# Add hemisphere names to the list
for name in names:
    hemisphere_names.append(name.text.strip('Enhanced'))
       
hemisphere_names
```
